```python
df_train['Date']=df_train['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df_train['Month']=df_train['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df_train['Year']=df_train['Date_of_Journey'].apply(lambda x:x.split('/')[2])                                                 
```

```python
df_train.head()
```

```python
df_train['Date']=df_train['Date'].astype(int)
df_train['Month']=df_train['Month'].astype(int)
df_train['Year']=df_train['Year'].astype(int)
```

```python
df_train.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df_train.head()
```

```python
df_train['Arrival_Time']=df_train['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df_train.head()
```

```python
df_train['Total_Stops'].unique()
```

```python
df_train['Total_Stops']=df_train['Total_Stops'].str.split(' ').str[0]
```

```python
ind=df_train[df_train['Total_Stops']=='non-stop'].index
for i in ind:
    df_train.loc[i,'Total_Stops']=0
```

```python
df_train.head()
```

```python
df_train['Total_Stops']=df_train['Total_Stops'].fillna(df_train['Total_Stops'].mode()[0])
```

```python
df_train['Total_Stops'].unique()
```

```python
df_train['Total_Stops']=df_train['Total_Stops'].astype(int)
```

```python
df_train.info()
```

```python
df_train['Additional_Info'].unique()
```

```python
df_train['Arrival_Hour']=df_train['Arrival_Time'].str.split(':').str[0]
```

```python
df_train['Arrival_Minute']=df_train['Arrival_Time'].str.split(':').str[1]
```

```python
df_train['Arrival_Hour']=df_train['Arrival_Hour'].astype(int)
```

```python
df_train['Arrival_Minute']=df_train['Arrival_Minute'].astype(int)
```

```python
df_train.drop('Arrival_Time',inplace=True,axis=1)
```

```python
df_train.head()
```

```python
df_train['Dep_Hour']=df_train['Dep_Time'].str.split(':').str[0]
df_train['Dep_Minute']=df_train['Dep_Time'].str.split(':').str[1]
```

```python
df_train.head()
```

```python
df_train['Dep_Hour']=df_train['Dep_Hour'].astype(int)
```

```python
df_train['Dep_Minute']=df_train['Dep_Minute'].astype(int)
```

```python
df_train.info()
```

```python

airlines_to_replace = ['Multiple carriers Premium economy', 'Jet Airways Business', 'Vistara Premium economy', 'Trujet']

df_train['Airline'] = df_train['Airline'].apply(lambda x: 'other' if x in airlines_to_replace else x)
```

```python
df_train['Airline'].unique()
```

```python
 df_train['Airline'].value_counts()
```

```python
Airlines=df_train[['Airline']]
Airline=pd.get_dummies(Airlines,drop_first=True)
Airline=Airline.astype(int)
Airline.head()
```

```python
df_train['Source'].value_counts()
```

```python
Sources=df_train[['Source']]
Source=pd.get_dummies(Sources,drop_first=True)
Source=Source.astype(int)
Source.head()
```

```python
Destinations=df_train[['Destination']]
Destination=pd.get_dummies(Destinations,drop_first=True)
Destination=Destination.astype(int)
Destination.head()
```

```python
df_train=pd.concat([df_train,Source,Destination,Airline],axis=1)
```
